Importing dependancies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.datasets import load_breast_cancer
```

Data Collection and Preprocessing

```python
df=pd.read_csv('breast-cancer_data.csv')
cancer=load_breast_cancer()
```

```python
df=df.drop('Unnamed: 32',axis=1)
df=df.drop('diagnosis',axis=1)
df=df.drop('id',axis=1)
df['label']=cancer.target
df
```

Data-Cleaning (Check for null Values)


```python
null_cols = df.isnull().any(axis=0)
null_cols = null_cols[null_cols == True]
print(null_cols.index.tolist())
```

Spliting datasets

```python
y=df['label']
X=df.drop('label',axis=1)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

Building Model and Model train

```python
model = LogisticRegression()
model.fit(X_train,y_train)
```

Predicting Values and Finding Accuracy

```python
y_pred = model.predict(X_test)
y_pred1 = model.predict(X_train)
```

```python
accuracy = accuracy_score(y_test, y_pred)
accuracy
```
