```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import LabelEncoder
```

```python
df_train = pd.read_csv(r"C:\Users\swast\Downloads\spaceship-titanic\train.csv")
df_test = pd.read_csv(r"C:\Users\swast\Downloads\spaceship-titanic\test.csv")
```

# Pre-Processing 

```python
df_train = df_train.drop(['PassengerId','Name','Cabin'],axis=1)
df_test = df_test.drop(['PassengerId','Name','Cabin'],axis=1)
```

```python
df_train
```

```python
df_test
```

```python
df_train.isnull().sum()
```

```python
#TRAIN
df_train['RoomService']=df_train['RoomService'].fillna(0.0)
df_train['FoodCourt']=df_train['FoodCourt'].fillna(0.0)
df_train['ShoppingMall']=df_train['ShoppingMall'].fillna(0.0)
df_train['Spa']=df_train['Spa'].fillna(0.0)
df_train['VRDeck']=df_train['VRDeck'].fillna(0.0)
```

```python
df_train['Age']=df_train['Age'].fillna(df_train['Age'].mean())
df_train['HomePlanet']=df_train['HomePlanet'].fillna('Earth')
df_train['CryoSleep']=df_train['CryoSleep'].fillna('False')
df_train['Destination']=df_train['Destination'].fillna('TRAPPIST-1e')
df_train['VIP']=df_train['VIP'].fillna('False')
```

```python
df_train['HomePlanet']=df_train['HomePlanet'].astype(str)
df_train['CryoSleep']=df_train['CryoSleep'].astype(str)
df_train['Destination']=df_train['Destination'].astype(str)
df_train['VIP']=df_train['VIP'].astype(str)
```

```python
#TEST
df_test['RoomService']=df_test['RoomService'].fillna(0.0)
df_test['FoodCourt']=df_test['FoodCourt'].fillna(0.0)
df_test['ShoppingMall']=df_test['ShoppingMall'].fillna(0.0)
df_test['Spa']=df_test['Spa'].fillna(0.0)
df_test['VRDeck']=df_test['VRDeck'].fillna(0.0)
```

```python
df_test['Age']=df_test['Age'].fillna(df_test['Age'].mean())
df_test['HomePlanet']=df_test['HomePlanet'].fillna('Earth')
df_test['CryoSleep']=df_test['CryoSleep'].fillna('False')
df_test['Destination']=df_test['Destination'].fillna('TRAPPIST-1e')
df_test['VIP']=df_test['VIP'].fillna('False')
```

```python
df_test['HomePlanet']=df_test['HomePlanet'].astype(str)
df_test['CryoSleep']=df_test['CryoSleep'].astype(str)
df_test['Destination']=df_test['Destination'].astype(str)
df_test['VIP']=df_test['VIP'].astype(str)
```

# Encoding

```python
encoder = LabelEncoder()
```

```python
#TRAIN
df_train['Encoded_HomePlanet']=encoder.fit_transform(df_train['HomePlanet'])
df_train['Encoded_CryoSleep']=encoder.fit_transform(df_train['CryoSleep'])
df_train['Encoded_Destination']=encoder.fit_transform(df_train['Destination'])
df_train['Encoded_VIP']=encoder.fit_transform(df_train['VIP'])
```

```python
#TEST
df_test['Encoded_HomePlanet']=encoder.fit_transform(df_test['HomePlanet'])
df_test['Encoded_CryoSleep']=encoder.fit_transform(df_test['CryoSleep'])
df_test['Encoded_Destination']=encoder.fit_transform(df_test['Destination'])
df_test['Encoded_VIP']=encoder.fit_transform(df_test['VIP'])
```

```python
df_train=df_train.drop(['HomePlanet','CryoSleep','Destination','VIP'],axis=1)
```

```python
df_test=df_test.drop(['HomePlanet','CryoSleep','Destination','VIP'],axis=1)
```
